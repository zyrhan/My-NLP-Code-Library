# Type 1: Count based

## Part 1.1:TF based 

### 1.1.1 TF: Term frequency

```python
### example data:
texts = [
    "The cat sat on the hat",
    "The dog ate the cat and the hat",
    "Dog and cat are good friends"
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
tf = CountVectorizer()
featuresTf = tf.fit_transform(texts)
pd.DataFrame(featuresTf.todense(),columns=tf.get_feature_names())
```

### 1.1.2 TF-IDF: term frequency inverse document frequency

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
tfidf = TfidfVectorizer()
features = tfidf.fit_transform(texts)
pd.DataFrame(features.todense(),columns=tfidf.get_feature_names())
```

## Part 1.2: SVD embedding: Co-occuranced based 

```python
uniqueTf = tfidf.get_feature_names()
len(uniqueTf)
```

```python
coMatrix = np.random.randint(3, size=(4, 11))
pd.DataFrame(columns=uniqueTf,data=coMatrix)  
```

```python
u, s, vt  = np.linalg.svd(coMatrix,full_matrices=False)
```

```python
print(f"shape of SVD component: {u.shape},{s.shape},{vt.shape}")
```

```python
pd.DataFrame(u@np.diag(s)@vt)
```

```python
from glove import Corpus, Glove
# creating a corpus object
corpus = Corpus() 
#training the corpus to generate the co occurence matrix which is used in GloVe
corpus.fit(texts, window=10)
#creating a Glove object which will use the matrix created in the above lines to create embeddings
#We can set the learning rate as it uses Gradient Descent and number of components
glove = Glove(no_components=50, learning_rate=0.05)
 
glove.fit(corpus.matrix, epochs=100, no_threads=4, verbose=True)
glove.add_dictionary(corpus.dictionary)
# glove.save('glove.model')
```

```python
glove.dictionary
```

```python
glove.word_vectors[glove.dictionary['trump']]
```

```python
glove.most_similar("trump",number=10)
```

```python
glove.most_similar("share",number=10)
```
