### Suppliment of VADER <br>

Summary:  The compound score is computed by summing the valence scores of each word in the lexicon, adjusted according to the rules, and then normalized to be between -1 (most extreme negative) and +1 (most extreme positive). This is the most useful metric if you want a single unidimensional measure of sentiment for a given sentence. Calling it a 'normalized, weighted composite score' is accurate.
1. positive sentiment: compound score >= 0.05
2. neutral sentiment: (compound score > -0.05) and (compound score < 0.05)
3. negative sentiment: compound score <= -0.05


```python
### Effect of punctuation
print(analyzer.polarity_scores("The food here is good!"))
print(analyzer.polarity_scores("The food here is good!!"))
print(analyzer.polarity_scores("The food here is good!!!"))
```

```python
### Effect of capitalization
print(analyzer.polarity_scores("The food here is good!"))
print(analyzer.polarity_scores("The food here is GOOD!"))
```

```python
### Effect of adjective
print(analyzer.polarity_scores("The food here is marginally good!"))
print(analyzer.polarity_scores("The food here is very good!"))
print(analyzer.polarity_scores("The food here is extremely good!"))
```

```python
### Effect of conjunction
print(analyzer.polarity_scores("The food here is good!, but service is bad"))
```

```python
### Effect of emojis
print(analyzer.polarity_scores('I am 😄 today'))
print(analyzer.polarity_scores('😊'))
print(analyzer.polarity_scores('😥'))
print(analyzer.polarity_scores('☹️'))
print(analyzer.polarity_scores('💋'))
print(analyzer.polarity_scores('💘'))
```

```python
### Effect of Slangs
print(analyzer.polarity_scores('It SUX!'))
print(analyzer.polarity_scores('It shits'))
print(analyzer.polarity_scores('It lol'))
```

## 2.2 TextBlob sentiment analysis

Supplement of TextBlob <br>

- Polarity is a float value within the range [-1.0 to 1.0] where 0 indicates neutral, +1 indicates a very positive sentiment and -1 represents a very negative sentiment.

- Subjectivity is a float value within the range [0.0 to 1.0] where 0.0 is very objective and 1.0 is very subjective. Subjective sentence expresses some personal feelings, views, beliefs, opinions, allegations, desires, beliefs, suspicions, and speculations where as Objective sentences are factual.

# Part 3: Modulize

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer as vader_analyzer
from textblob import TextBlob

class featureExtractor:

    def vader_score(self,text):
        vader = vader_analyzer()
        featureDic = vader.polarity_scores(text)
        return featureDic

    def text_blob_score(self,text):
        text_blob = TextBlob(text)
        featureDic = {"plarity":blob.sentiment.polarity,"subjectivity":blob.sentiment.subjectivity}
        return featureDic

    def get_sent(self,text):
        featureDic1 = self.vader_score(text)
        featureDic2 = self.text_blob_score(text)
        return {**featureDic1,**featureDic2}
    
class featurePipeline:
    def __init__(self,texts):
        self.texts = texts
        self.extractor = featureExtractor()
        
    def run(self,workers=6):
        with ProcessPoolExecutor(max_workers=workers) as executor:
            res = executor.map(self.extractor.get_sent, self.texts)         
        return list(res)
```
